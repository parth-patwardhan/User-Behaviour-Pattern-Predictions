```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import torch.nn.functional as F

data_path = 'data/reddit.csv'
df = pd.read_csv(data_path)




user_encoder = LabelEncoder()
item_encoder = LabelEncoder()

df['user_id'] = user_encoder.fit_transform(df['user_id'])
df['timestamp'] = item_encoder.fit_transform(df['timestamp'])

edges = df[['user_id', 'timestamp']].values.T


edge_attr = df['item_id'].values


num_users = len(user_encoder.classes_)
num_items = len(item_encoder.classes_)


user_features = torch.randn(num_users, 16)
item_features = torch.randn(num_items, 16)

x = torch.cat([user_features, item_features], dim=0)
edge_index = torch.tensor(edges, dtype=torch.long)
edge_attr = torch.tensor(edge_attr, dtype=torch.float)

data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr)

train_edges, test_edges = train_test_split(edges.T, test_size=0.4, random_state=42)

train_data = Data(x=x, edge_index=torch.tensor(train_edges.T, dtype=torch.long))

class GCNLinkPrediction(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(GCNLinkPrediction, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        # GCN layers
        x = F.relu(self.conv1(x, edge_index))
        x = self.conv2(x, edge_index)
        return x

model = GCNLinkPrediction(in_channels=16, hidden_channels=64, out_channels=32)  # 32 is the embedding size for each node
optimizer = optim.Adam(model.parameters(), lr=0.01)
criterion = nn.BCEWithLogitsLoss()

def train():
    model.train()
    optimizer.zero_grad()

    out = model(train_data.x, train_data.edge_index)

    source_embeddings = out[train_data.edge_index[0]]
    target_embeddings = out[train_data.edge_index[1]]

    edge_predictions = (source_embeddings * target_embeddings).sum(dim=-1)

    labels = torch.ones_like(edge_predictions)

    loss = criterion(edge_predictions, labels)
    loss.backward()
    optimizer.step()
    return loss

def test():
    model.eval()
    with torch.no_grad():
        out = model(train_data.x, train_data.edge_index)

        source_embeddings = out[train_data.edge_index[0]]
        target_embeddings = out[train_data.edge_index[1]]

        edge_predictions = (source_embeddings * target_embeddings).sum(dim=-1)

        pred = (edge_predictions > 0).float()

        correct = pred.eq(torch.ones_like(pred)).sum().item()
        accuracy = correct / len(pred)

    return accuracy

# Training and evaluation
for epoch in range(1oo):
    loss = train()
    acc = test()
    print(f'Epoch {epoch + 1}: Loss: {loss:.4f}, Accuracy: {acc:.4f}')
```
